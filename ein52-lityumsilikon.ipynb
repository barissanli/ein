```python
tumuzunpozisyonlar="Total Reportable Positions-Long (Old)"
tumkisapozisyonlar="Total Reportable Positions-Short (All)"
spekuzunpozisyonlar="Noncommercial Positions-Long (All)"
spekkisapozisyonlar="Noncommercial Positions-Short (All)"
ticarikisapozisyonlar="Commercial Positions-Short (All)"
ticariuzunpozisyonlar="Commercial Positions-Long (All)"
```

# Önce Lityumdaki duruma bakalım

```python
# Set the theme to resemble ggplot2 
sns.set_theme(style="whitegrid") 

emtiaismi="LITHIUM HYDROXIDE - COMMODITY EXCHANGE INC."

# Create a figure with size 12x6 
plt.figure(figsize=(12, 6))
vs=df[df["Market and Exchange Names"]==emtiaismi][tumkisapozisyonlar].values
vl=df[df["Market and Exchange Names"]==emtiaismi][tumuzunpozisyonlar].values
dts=df[df["Market and Exchange Names"]==emtiaismi]['Dates'].values
plt.plot(dts,vs,label="Kısa Pozisyonlar")
plt.plot(dts,vl,label="Uzun Pozisyonlar")
#a
plt.xticks(rotation=90);
plt.title("Lityum Pozisyonları - Tüm Pozisyonlar")
plt.legend()
#plt.plot(df["Total Reportable Positions-Long (Old)"].values)
```

```python

# Create a figure with size 12x6 
plt.figure(figsize=(12, 6))
vs=df[df["Market and Exchange Names"]==emtiaismi][spekkisapozisyonlar].values
vl=df[df["Market and Exchange Names"]==emtiaismi][spekuzunpozisyonlar].values
vs2=df[df["Market and Exchange Names"]==emtiaismi][ticarikisapozisyonlar].values
vl2=df[df["Market and Exchange Names"]==emtiaismi][ticariuzunpozisyonlar].values
dts=df[df["Market and Exchange Names"]==emtiaismi]['Dates'].values
plt.plot(dts,vs,label="Spek-Kısa Pozisyonlar")
plt.plot(dts,vl,label="Spek-Uzun Pozisyonlar")
plt.plot(dts,vs2,label="Ticari-Kısa Pozisyonlar")
plt.plot(dts,vl2,label="Ticari-Uzun Pozisyonlar")
#a
plt.xticks(rotation=90);
plt.title("Lityum Pozisyonları")
plt.legend()
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation=90);
plt.plot(dts,vl-vs,label="Net posizyonlar",color="#0000AA")
#plt.plot(dts,vs,label="Kısa posizyonlar",color="#990000")
plt.title("Lityum Net Pozisyonlar")
plt.legend()
```
